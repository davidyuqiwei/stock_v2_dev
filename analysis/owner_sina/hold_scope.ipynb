```python
def get_data_owner():
    return f"""select * from m_owner_stock_name_price_hs300 """

conn = CommonScript.connect_to_db("test.db")
df1 = pd.read_sql_query(get_data_owner(),conn)
ownler_list = df1[["owner_name","stock_name"]].drop_duplicates()
ownler_list_v1 = ownler_list.values.tolist()


owner_name_list_v2 = []
stock_name_v2 = []
slope_v2 = []
for i in ownler_list_v1:
    # print(i)
    df2 = df1[(df1["owner_name"]==i[0])&(df1["stock_name"]==i[1])]
    
    if df2.shape[0]>5 and df2['update_date'].max()[0:4]=='2024':
        df2 = df2.sort_values("update_date").reset_index()
        slope, intercept = np.polyfit(df2.index, df2['hold_num']/df2['hold_num'].max(), 1)
        owner_name_list_v2.append(i[0])
        stock_name_v2.append(i[1])
        slope_v2.append(slope)
```

```python
df3 = pd.DataFrame([owner_name_list_v2,stock_name_v2,slope_v2]).T
df3.columns = ["owner_name","stock_name","slope"]
df3.sort_values("slope",ascending=False).head(20)
```
